```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
knn = KNeighborsClassifier(n_neighbors=1)

path = 'ionosphere.data'
df = pd.read_csv(path, header= None)
```

```python
X = df.iloc[:, [i for i in range(0, 34)]]
y = df.iloc[:, [34]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
knn.fit(X_train, y_train)
```

```python
k = knn.score(X_test, y_test)
print(k)
```

```python
p = knn.predict([[11,0,0.94490,-0.49311,1,-0.03692,0.98898,-0.87052,0.90083,0.66942,1,-0.10104,1,-0.12493,1,-0.15017,1,-0.17681,1,-0.20491,1,-0.23452,1,-0.26571,1,-0.29852,1,-0.33304,1,-0.36931,1,-0.40740,1,-0.44739]])
```

```python
if k > 0.8:
    print('属于' + p + '类')
else:
    print('测试数据集不合格')
```
