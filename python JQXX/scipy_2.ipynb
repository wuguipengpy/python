```python

import numpy as np 
arr = [52, 55, 49, 43, 61, 53, 48, 46, 51, 53, 50, 42, 58, 51, 45]
#求均值
arr_mean = np.mean(arr)
#求方差
arr_var = np.var(arr)
#求标准差
arr_std = np.std(arr,ddof=1)
print("平均值为：%f" % arr_mean)
print("方差为：%f" % arr_var)
print("标准差为:%f" % arr_std)
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

rv = st.poisson(50)
num_years = [1,1,0,1,1,0,1,1,1,2,1,2,0,1,0,0,1,0,0,1]
x = range(42, 62)
plt.bar(np.array(x)-.4, num_years, label='Observed instances')
plt.plot(x, sum(num_years)*rv.pmf(x), ls='dashed', 
        lw=2, c='r', label='Poisson distribution\n$(\lambda=50)$')
plt.xlim([40, 70])       
plt.ylim([0, 2])
plt.xlabel('# of mass shootings in a year')
plt.ylabel('# of years')
plt.legend(loc='best')
plt.show()
```

```python
from math import exp, factorial

def poison_pdf(lamb, k):
    return lamb ** k * exp(-lamb) / factorial(k)

def poison_cdf(lamb, k):
    return sum([poison_pdf(lamb, i) for i in range(k + 1)])
s = ('一天刚好卖{0}杯豆浆的概率为{1:.2f}%,'
    '不超过{2}杯豆浆的概率为{3:.2f}%')
for i in range(100):
    print(s.format(i, poison_pdf(50, i) * 100,
                  i, poison_cdf(30, i) * 100))
```

```python
fig,ax = plt.subplots(1,1)
mu = 50
#平均值, 方差, 偏度, 峰度
mean,var,skew,kurt = st.poisson.stats(mu,moments='mvsk')
print(mean,var,skew,kurt)
#ppf:累积分布函数的反函数。q=0.01时，ppf就是p(X<x)=0.01时的x值。
x = np.arange(st.poisson.ppf(0.01, mu),st.poisson.ppf(0.99, mu))
ax.plot(x, st.poisson.pmf(x, mu),'o')
plt.title(u'poisson分布概率质量函数')
plt.show()
```
