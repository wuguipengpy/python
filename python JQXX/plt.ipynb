```python
x = np.linspace(-math.pi, math.pi, num = 256) #for循环效率慢改用np数组
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
axis = fig.add_subplot(211)

axis.set_ylim(-0.5 * math.pi, 0.5 * math.pi)
axis.grid(True)
plt.plot(x, y, 'r--',label = 'sin(x)')
plt.legend(loc = 'best')

axis = fig.add_subplot(212)
plt.plot(x, z, 'b--', linewidth = 4, label = 'cos(x)')
plt.legend(loc = 'best')

plt.savefig('D:\PYTP\plot-a.jpg')
plt.show()
```

```python
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-,', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color = 'black')
ax.axis('equal')
    
ax.legend(lines[:2], ['line A', 'line B'],
         loc = 'upper right', frameon = False)

from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
            loc = 'lower right', frameon = False)

ax.add_artist(leg)
plt.savefig('D:\PYTP\plot.jpg')
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-math.pi, math.pi, num = 48)
y = np.sin(x + 0.05 * np.random.standard_normal(len(x)))
y_error = 0.1 * np.random.standard_normal(len(x))

fig = plt.figure()
axis = fig.add_subplot(111)
axis.set_ylim(-0.5 * math.pi, 0.5 * math.pi)

plt.plot(x, y, 'r--',label='sin(x)')
plt.errorbar(x, y, yerr = y_error)

plt.legend(loc = 'best')
plt.show()
```

```python
nbPointers = 50

x = np.random.standard_normal(nbPointers)
y = np.random.standard_normal(nbPointers)


color = np.random.rand(nbPointers)#随机产生颜色

#plt.scatter(x, y)

area = (30 * np.random.rand(nbPointers)) ** 2 #随机产生半径
plt.scatter(x, y, s = area, c = color, alpha = 0.5) #alpha设置透明度，越靠近0，透明度越高

plt.show()
```

```python
data = []
for _ in range(5):   #用for循环打印5个盒须图
    mu = 10 * np.random.random_sample()
    sigma = 2 * np.random.random_sample() + 0.1
    data.append(np.random.normal(mu, sigma, 256))
    
plt.boxplot(data)
plt.show()
```
