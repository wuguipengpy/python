```python
import pandas as pd

p = {'数值特征':[5, 6, 7, 8, 9], '类型特征':['西瓜', '香蕉', '橘子', '苹果', '葡萄']}
fruits = pd.DataFrame(p)
fruits
```

```python
import numpy as np
import matplotlib.pyplot as plt
rnd = np.random.RandomState(38)
x = rnd.uniform(-5, 5, size = 50)

y_no_noise = (np.cos(6*x) + x)
X = x.reshape(-1, 1)
y = (y_no_noise + rnd.normal(size = len(x)))/2
plt.plot(X, y, 'o', c = 'r')
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor

line = np.linspace(-5, 5, 1000, endpoint=False).reshape(-1, 1)

mlpr = MLPRegressor().fit(X, y)
knr = KNeighborsRegressor().fit(X, y)

plt.plot(line, mlpr.predict(line), label = 'MLP')
plt.plot(line, knr.predict(line), label = 'KNN')
plt.plot(X, y, 'o', c = 'r')
plt.legend(loc = 'best')
plt.show()
```

```python
bins = np.linspace(-5, 5, 11)
target_bin = np.digitize(X, bins = bins)
print('装箱数据范围: \n{}'.format(bins))
print('\n前十个数据点的特征值: \n{}'.format(X[:10]))
print('\n前十个数据点所在的箱子:\n{}'.format(target_bin[:10]))
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder(sparse=False)
onehot.fit(target_bin)
X_in_bin = onehot.transform(target_bin)
print('装箱后的数据形态:{}'.format(X_in_bin.shape))
print('\n装箱后前十个数据点:{}'.format(X_in_bin[:10]))
```

```python
new_line = onehot.transform(np.digitize(line, bins = bins))

mlpr = MLPRegressor().fit(X_in_bin, y)
knr = KNeighborsRegressor().fit(X_in_bin, y)

plt.plot(line, mlpr.predict(new_line), label = 'MLP')
plt.plot(line, knr.predict(new_line), label = 'KNN')
plt.plot(X, y, 'o', c = 'r')
plt.legend(loc = 'best')
plt.show()
```

```python
#升维前
print(X_in_bin.shape)
```

```python
#升维后
X_stack = np.hstack([X, X_in_bin])
print(X_stack.shape)
```

```python
line_stack = np.hstack([line, new_line])

mplr_interact = MLPRegressor()
mplr_interact.fit(X_stack, y)
plt.plot(line, mplr_interact.predict(line_stack), label = 'MLP for interaction')
plt.ylim(-4, 4)
for vline in bins:
    plt.plot([vline, vline], [-5, 5], ':', c = 'k')
plt.legend(loc = 'best')
plt.plot(X, y, 'o', c = 'r')
plt.show()
```

```python
X_multi = np.hstack([X_in_bin, X*X_in_bin])
print(X_multi.shape)
print(X_multi[0])
```

```python
mplr_multi = MLPRegressor()
mplr_multi.fit(X_multi, y)
line_multi = np.hstack([new_line, line * new_line])
plt.plot(line, mplr_multi.predict(line_multi), label = 'MLP Regessor')
for vline in bins:
    plt.plot([vline, vline], [-5, 5], ':', c = 'gray')
plt.plot(X, y, 'o', c = 'r')
plt.legend(loc = 'lower right')
plt.show()
```
