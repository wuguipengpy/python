```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap

line = np.linspace(-5, 5, 200)

plt.plot(line, np.tanh(line), label = 'tanh')
plt.plot(line, np.maximum(line, 0), label = 'relu')

plt.legend(loc = 'best')

plt.show()
```

```python
wine = load_wine()

X = wine.data[:, :2]
y = wine.target

X_rtain, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

mlp = MLPClassifier(solver='lbfgs')
mlp.fit(X_rtain, y_train)



```

```python
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

x_min, x_max = X_rtain[:, 0].min() - 1, X_rtain[:, 0].max() + 1
y_min, y_max = X_rtain[:, 1].min() - 1, X_rtain[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max),
                     np.arange(y_min, y_max))

Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

plt.scatter(X[:, 0], X[:, 1], c = y, edgecolors='k', s = 60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
mlp_20 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10])
mlp_20.fit(X_rtain, y_train)
Z = mlp_20.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

plt.scatter(X[:, 0], X[:, 1], c = y, edgecolors='k', s = 60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
mlp_2L = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10,10])
mlp_2L.fit(X_rtain, y_train)
Z = mlp_2L.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

plt.scatter(X[:, 0], X[:, 1], c = y, edgecolors='k', s = 60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
mlp_tanh = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10],  activation='tanh')
mlp_tanh.fit(X_rtain, y_train)
Z = mlp_tanh.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

plt.scatter(X[:, 0], X[:, 1], c = y, edgecolors='k', s = 60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
mlp_alpha = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10],  activation='tanh', alpha=1)
mlp_alpha.fit(X_rtain, y_train)
Z = mlp_alpha.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

plt.scatter(X[:, 0], X[:, 1], c = y, edgecolors='k', s = 60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()

```
