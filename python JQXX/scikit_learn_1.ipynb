```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
df =pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None) #header去掉头部，只留数据
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,[i for i in range(1, 14)]], df.iloc[:, [0]], random_state=0)
knn.fit(X_train, y_train)
a = [12.1, 2.98, 3.12, 19.2, 95.2, 1.05, 2.58, 0.59, 1.52, 6.3, 1.18, 3.78, 853]
k = knn.predict([a])
print(k)
```
