```python
import matplotlib.pyplot as plt
import math
plt.plot([0, 2], [0, 2])
plt.plot([2,4], [2, 2])
plt.plot([4,6], [2, -2])
plt.plot([6, 8], [-2, -2])
plt.plot([8, 10], [-2, 0])
plt.show()
```

```python
nbSamples = 256
xRange = (-math.pi, math.pi)

x, y = [], []

for n in range(nbSamples):
    k = (n + 0.5) / nbSamples
    x.append(xRange[0] + (xRange[1] - xRange[0]) * k)
    y.append(math.sin(x[-1]))
    
plt.plot(x, y)
plt.show()
```
